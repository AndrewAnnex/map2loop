```python
''' 
In this example we generate synthetic scattered data with added noise
and then fit it with a smoothed RBF interpolant. The interpolant in
this example is equivalent to a thin plate spline.
'''
import numpy as np
from rbf.interpolate import RBFInterpolant
import matplotlib.pyplot as plt
np.random.seed(1)

# observation points
x_obs = np.random.random((100, 2)) 
#display(x_obs)
# values at the observation points
u_obs = np.sin(2*np.pi*x_obs[:, 0])*np.cos(2*np.pi*x_obs[:, 1]) 
u_obs += np.random.normal(0.0, 0.3, 100)
# create a thin-plate spline interpolant, where the data is assumed to
# be noisy

subset=5000
methods=(
#'phs8',
#'phs7',
#'phs6',
#'phs5',
#'phs4',
#'phs3',
#'phs2',
'phs1',
'mq'#,
#'imq',
#'iq',
#'ga',
#'exp',
#'se',
#'mat32',
#'mat52'
)
import os
print(os.getcwd())
contacts=np.genfromtxt('raw_contacts.csv',delimiter=',',skip_header=1,dtype='f8')
x_obs=contacts[0:subset,0:2]
display(x_obs)
u_obs=contacts[0:subset,4]
display(u_obs)
print(np.amin(contacts[0:,0]))
print(np.amax(contacts[0:,0]))
print(np.amin(contacts[0:,1]))
print(np.amax(contacts[0:,1]))

xmin=np.amin(contacts[0:,0])
xmax=np.amax(contacts[0:,0])
ymin=np.amin(contacts[0:,1])
ymax=np.amax(contacts[0:,1])

display(x_obs.shape,u_obs.shape)
```

```python
for method in methods:
    for eps in np.arange(1,0,-.1):
    #method='phs2'
        sigma=.5
        print(method,sigma,eps)
        I = RBFInterpolant(x_obs, u_obs, eps=eps,sigma=sigma, phi=method, order=1)
        #print('done')
        # create the interpolation points, and evaluate the interpolant
        x1, x2 = np.linspace(xmin, xmax, 100), np.linspace(ymin, ymax, 100)
        x_itp = np.reshape(np.meshgrid(x1, x2), (2, 100*100)).T
        u_itp = I(x_itp) 
        # plot the results
        plt.tripcolor(x_itp[:, 0], x_itp[:, 1], u_itp, vmin=0, vmax=1, cmap='viridis')
        plt.scatter(x_obs[:, 0], x_obs[:, 1], s=100, c=u_obs, vmin=0, vmax=1,
                    cmap='viridis', edgecolor='k')
        plt.xlim((xmin, xmax))
        plt.ylim((ymin, ymax))
        plt.colorbar()
        plt.tight_layout()
        #plt.savefig('../interpolate.a.png')
        plt.show()
```
